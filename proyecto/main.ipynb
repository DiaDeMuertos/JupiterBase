```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import simplefilter

simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

```python
df = pd.read_csv('data/une_sample.csv', encoding='utf-8')
df.head(2)
```

```python
df.drop(['nombre_ruta'], axis=1, inplace=True)
df.head()
```

```python
df.groupby(['vehiculo_id']).count()
```

```python
df['fecha'] = pd.DatetimeIndex(df['fecha']).tz_convert(tz='MST')
df['hora'] = pd.DatetimeIndex(df['fecha']).hour
df['minutos'] = pd.DatetimeIndex(df['fecha']).minute

df.head(2)
```

```python
cond_a = df['hora'] >= 17
cond_b = df['hora'] <= 18
cond_c = df['vehiculo_id'] == 150

temp = df[cond_a & cond_b & cond_c].copy()

temp.sort_values(by=['fecha','vehiculo_id'])
```
